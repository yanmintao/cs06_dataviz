```
import altair as alt
from vega_datasets import data
cars = data.cars()
```

```
points = alt.Chart(cars).mark_point().encode(
    x='Miles_per_Gallon:Q',
    y='Horsepower:Q',
    color='Origin:N'
)

bar1 = alt.Chart(cars).mark_bar().encode(
    y='Origin:N',
    x='count(Origin):Q',
    color='Origin'
)

bar2 = alt.Chart(cars).mark_bar().encode(
    x='Cylinders:Q',
    y='count(Name):Q',
    color='Origin'
)
points & bar1 |bar2
```

```
# Add the interactive window
brush = alt.selection(type='interval') # select an area

points = alt.Chart(cars).mark_point().encode(
    x='Miles_per_Gallon:Q',
    y='Horsepower:Q',
    color='Origin:N'
).add_selection(brush) # add window to points chart

bar1 = alt.Chart(cars).mark_bar().encode(
    y='Origin:N',
    x='count(Origin):Q',
    color='Origin'
).transform_filter(brush) # use brush as the filter

hist = alt.Chart(cars).mark_bar().encode(
    x = alt.X('Weight_in_lbs:Q', bin=True),
    y='count()'
).transform_filter(brush)

bar2 = alt.Chart(cars).mark_bar().encode(
    x='Cylinders:Q',
    y='count(Name):Q',
    color='Origin'
).transform_filter(brush)
points & bar1 | hist & bar2 # Display 3 charts connected by brush window
```

```
brush = alt.selection(type='interval')

points = alt.Chart(cars).mark_point().encode(
    x='Miles_per_Gallon:Q',
    y='Horsepower:Q',
    color=alt.condition(brush, 'Origin:N', alt.value('lightgrey')) # grey outside the window
).add_selection(brush)

bar = alt.Chart(cars).mark_bar().encode(
    y='Origin:N',
    x='count(Origin):Q',
    color='Origin'
).transform_filter(brush)

hist = alt.Chart(cars).mark_bar().encode(
    x = alt.X('Weight_in_lbs:Q', bin=True),
    y='count()'
).transform_filter(brush)

combined_chart = points & bar | hist
```
