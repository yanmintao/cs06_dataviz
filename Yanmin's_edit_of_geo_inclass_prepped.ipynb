# Altair

```
import altair as alt
from vega_datasets import data
```

```
source = data.population_engineers_hurricanes()
source
```

```
states = alt.topo_feature(data.us_10m.url, 'states')
states
```

```

alt.Chart(states).mark_geoshape().encode(
            alt.Color('population', type='quantitative'),
    ).transform_lookup(
        lookup='id',
        from_=alt.LookupData(source, 'id', ['state', 'population', 'engineers'])
    ).properties(
        width=500,
        height=400
    ).project(
        type='albersUsa' # Albers USA map style 
    )
```

```
# Map the Hurricanes
alt.Chart(states).mark_geoshape().encode(
            alt.Color('hurricanes', type='quantitative'),
    ).transform_lookup(
        lookup='id',
        from_=alt.LookupData(source, 'id', ['state','population','hurricanes'])
    ).properties(
        width=500,
        height=400
    ).project(
        type='albersUsa' # Albers USA map style 
    )
```

```
# Geo shape, interactive, map the engineers population
alt.Chart(states).mark_geoshape().encode(
        alt.Color('engineers', type='quantitative'),
        # tooltip=['state:N', 'population:Q', 'engineers:Q']
        tooltip=['state:N',
                 alt.Tooltip('population:Q', format=',.0f'),
                 alt.Tooltip('engineers:Q', format='.1%')]
    ).transform_lookup(
        lookup='id',
        from_=alt.LookupData(source, 'id', ['state', 'population', 'engineers'])
    ).properties(
        width=500,
        height=300
    ).project(
        type='albersUsa'
    )
```
